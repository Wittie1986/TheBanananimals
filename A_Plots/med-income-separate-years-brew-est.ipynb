```python
# Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
# Load in CSV

brew_df = pd.read_csv("census-brewery-region-data.csv")
brew_df.head()
```

```python
# Removing Duplicates

brew_df.drop_duplicates(subset = "Brewery Name", keep='first', inplace = True)
brew_df.head()
```

```python
brew_df = brew_df[(brew_df["Year Established"] >= 2010) & (brew_df["Year Established"] <= 2017)]
brew_df.head()
```

```python
cm = plt.cm.get_cmap('copper') # Colour map

brew_df.plot(kind="scatter", x="Year Established", y="Median 2010", color=cm(0.2))

plt.title("2010 Median Income vs. Brewery's Year Established")
plt.xlabel("Brewery Year Established")
plt.ylabel("Median Income ($)")
plt.ylim(ymax=90000)
```

```python
plt.savefig("Images/ByYear/2010-Incomes-vs-Brewery-Year-Established.png")
```
