```python
# Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
# Load in CSV

brew_df = pd.read_csv("census-brewery-region-data.csv")
brew_df.head()
```

```python
# Removing Duplicates

brew_df.drop_duplicates(subset = "Brewery Name", keep='first', inplace = True)
brew_df.head()
```

```python
brew_df = brew_df[(brew_df["Year Established"] >= 2010) & (brew_df["Year Established"] <= 2017)]
brew_df.head()
```

```python
cm = plt.cm.get_cmap('copper') # Colour map

ax = brew_df.plot(kind="scatter", x="Year Established", y="2010 POP", figsize=(10,7.5), color=cm(0.2), label="2010");
brew_df.plot(kind="scatter", x="Year Established", y="2011 POP", color=cm(0.3), label="2011", ax=ax);
brew_df.plot(kind="scatter", x="Year Established", y="2012 POP", color=cm(0.4), label="2012", ax=ax);
brew_df.plot(kind="scatter", x="Year Established", y="2013 POP", color=cm(0.5), label="2013", ax=ax);
brew_df.plot(kind="scatter", x="Year Established", y="2014 POP", color=cm(0.6), label="2014", ax=ax);
brew_df.plot(kind="scatter", x="Year Established", y="2015 POP", color=cm(0.7), label="2015", ax=ax);
brew_df.plot(kind="scatter", x="Year Established", y="2016 POP", color=cm(0.8), label="2016", ax=ax);
brew_df.plot(kind="scatter", x="Year Established", y="2017 POP", color=cm(0.9), label="2017", ax=ax)

plt.title("Population per Year vs. Brewery's Year Established")
plt.xlabel("Brewery Year Established")
plt.ylabel("Population Size")
plt.ylim(ymax=2200000)
plt.legend(title="Population Year", ncol=4, loc=2)
```
