```python
# Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
# Load in CSV

brew_df = pd.read_csv("census-brewery-region-data.csv")
brew_df.head()
```

```python
# Removing Duplicates

brew_df.drop_duplicates(subset = "Brewery Name", keep='first', inplace = True)
brew_df.head()
```

```python
brew_df = brew_df[(brew_df["Year Established"] >= 2010) & (brew_df["Year Established"] <= 2017)]
brew_df.head()
```

```python
brew_closed_df = brew_df[(brew_df["Closed?"] == "Y")]
brew_closed_df.head()
```

```python
# Create Groups for Year Established
est_close_group = brew_closed_df.groupby('Year Established')

# Count how many times each Year Established appears
count_est_close = est_close_group['Year Established'].count()

count_est_close
```

```python
# Create a bar chart based off the group series

est_chart = count_est_close.plot(kind='line', colormap="copper_r")

# Set xlabel & ylabel & title

plt.title("Closed Breweries by Brewery's Year Established")
plt.xlabel("Year Established")
plt.ylabel("Number of Breweries Closed")


plt.show()
plt.tight_layout()
```
