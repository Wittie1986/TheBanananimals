```python
# Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
# Load in CSV

brew_df = pd.read_csv("census-brewery-region-data.csv")
brew_df.head()
```

```python
# Removing Duplicates

brew_df.drop_duplicates(subset = "Brewery Name", keep='first', inplace = True)
brew_df.head()
```

```python
brew_df = brew_df[(brew_df["Year Established"] >= 2010) & (brew_df["Year Established"] <= 2017)]
brew_df.head()
```

```python
# Create Groups for Regions
reg_group = brew_df.groupby('Region')

# Count how many times each Region appears
count_reg = reg_group['Region'].count()

count_reg
```

```python
# Create a bar chart based off the group series

reg_chart = count_reg.plot(kind='bar', colormap="copper_r")

# Set xlabel & ylabel & title

plt.title("Breweries per Region")
plt.xlabel("Region")
plt.ylabel("Number of Breweries")


plt.show()
plt.tight_layout()
```
