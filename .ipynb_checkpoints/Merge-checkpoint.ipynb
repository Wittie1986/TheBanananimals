```python
#Import Dependencies

import pandas as pd
import numpy as np
```

```python
#name CSVs

brew1 = "Output/brewery_data_10.csv"
brew2 = "Output/brewery_data_20.csv"
brew3 = "Output/brewery_data_30.csv"
brew4 = "Output/brewery_data_40.csv"
brew5 = "Output/brewery_data_50.csv"
brew6 = "Output/brewery_data_60.csv"
brew7 = "Output/brewery_data_70.csv"
brew8 = "Output/brewery_data_80.csv"
brew9 = "Output/brewery_data_90.csv"
brew100 = "Output/brewery_data_100.csv"
brew11 = "Output/brewery_data_110.csv"
```

```python
#read in named CSVs

brew10 = pd.read_csv(brew1)
brew20 = pd.read_csv(brew2)
brew30 = pd.read_csv(brew3)
brew40 = pd.read_csv(brew4)
brew50 = pd.read_csv(brew5)
brew60 = pd.read_csv(brew6)
brew70 = pd.read_csv(brew7)
brew80 = pd.read_csv(brew8)
brew90 = pd.read_csv(brew9)
brew_10 = pd.read_csv(brew100)
brew110 = pd.read_csv(brew11)

```

```python
#List of DFs
DFs = [brew10, brew20, brew30, brew40, brew50, brew60, brew70, brew80, brew90, brew_10, brew110]
```

```python
df = pd.concat(DFs, axis=0)
```

```python
df1 = df.reset_index(drop=True)
```

```python
df1.count()
```

```python
df1.dtypes
```

```python
df1
```

```python
df2 = df1.replace("['null']", np.NaN)
df2
```

```python
df_1 = df2.loc[df2["Country"] == "UNITED STATES"]
df_1
```

```python
BrewDF = df_1.reset_index(drop=True)
BrewDF
```

```python
brewdf = BrewDF.loc[BrewDF["Year Established"] == (2010)]
brewdf.head(1)
```

```python
brewdf1 = BrewDF.loc[BrewDF["Year Established"] == (2011)]
brewdf1.head(1)
```

```python
brewdf2 = BrewDF.loc[BrewDF["Year Established"] == (2012)]
brewdf2.head(1)
```

```python
brewdf3 = BrewDF.loc[BrewDF["Year Established"] == (2013)]
brewdf3.head(1)
```

```python
brewdf4 = BrewDF.loc[BrewDF["Year Established"] == (2014)]
brewdf4.head(1)
```

```python
brewdf5 = BrewDF.loc[BrewDF["Year Established"] == (2015)]
brewdf5.head(1)
```

```python
brewdf6 = BrewDF.loc[BrewDF["Year Established"] == (2016)]
brewdf6.head(1)
```

```python
brewdf7 = BrewDF.loc[BrewDF["Year Established"] == (2017)]
brewdf7.head(1)
```

```python
brewdf8 = BrewDF.loc[BrewDF["Year Established"] == (2018)]
brewdf8.head(1)
```

```python
DF1s = [brewdf, brewdf1, brewdf2, brewdf3, brewdf4, brewdf5, brewdf6, brewdf7, brewdf8]
```

```python
BreweryDF = pd.concat(DF1s, axis=0)
BreweryDF
```

```python
BreweryDFs = BreweryDF.reset_index(drop=True)
BreweryDFs
```

```python
BreweryDFs.to_csv("Output/BreweryData.csv", encoding="utf-8", index=False, header=True)
```

```python
Closed = BreweryDFs.loc[BreweryDFs["Closed?"] == "Y"]
Closed
```

```python
Closed.to_csv("Output/Closed.csv", encoding="utf-8", index=False, header=True)
```

```python
Brew = BreweryDFs.groupby(["State"])
```

```python
Brew_count = Brew.count()

Brew_count
```

```python
Brew_count.to_csv("Output/Count.csv", encoding="utf-8", index=False, header=True)
```
