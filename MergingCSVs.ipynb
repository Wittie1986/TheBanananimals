```python
# Import Dependencies
import pandas as pd
```

```python
# Define File Path for CSVs
brewery1_csv = "Output/brewery_data_part1.csv"
brewery2_csv = "Output/brewery_data_part2.csv"
brewery3_csv = "Output/brewery_data_part3.csv"
```

```python
# Read in CSVs
brewerydf_1 = pd.read_csv(brewery1_csv)
brewerydf_2 = pd.read_csv(brewery2_csv)
brewerydf_3 = pd.read_csv(brewery3_csv)
```

```python
# Merge CSVs
brewerydf = pd.merge(brewerydf_1, brewerydf_2, how="outer")
brewerydf.head()
```

```python
# Merge CSVs
brewerydf2 = pd.merge(brewerydf, brewerydf_3, how="outer")
brewerydf2.head()
```

```python
# Export new Merged CSV
brewerydf2.to_csv("Output/brewery_data_merged.csv",
                   encoding="utf-8", index=False, header=True)
```
